```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

### Dataset

```python
data = sns.load_dataset('tips')
data.head()
```

## Line plot

```python
x = data['day']  
y = data['tip']

fig, ax = plt.subplots()
fig.set_facecolor('black')
ax.set_facecolor('black')
for spines in ax.spines.values():
    spines.set_color('white')
ax.tick_params(axis='both',colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
sns.lineplot(x=x, y=y)
plt.show()
```

## Bar Plot 

```python
fig, ax = plt.subplots()
fig.set_facecolor('black')
ax.set_facecolor('black')
for spines in ax.spines.values():
    spines.set_color('white')
ax.tick_params(axis='both',colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
sns.barplot(x=x, y=y, color='cyan', width=0.2)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_facecolor('black')
ax.set_facecolor('black')
for spines in ax.spines.values():
    spines.set_color('white')
ax.tick_params(axis='both',colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

sns.barplot(x=data['day'], y=data['total_bill'],  label='total_bill', color='cyan', width=0.4)
sns.barplot(x=data['day'], y=data['tip'], bottom=data['total_bill'], label='tip', color='orange', width=0.4)
plt.show()
```

## HeatMap graph 

```python
sns.heatmap(data[['total_bill','tip']].corr(), annot=True, cmap='coolwarm')
```

## Count Plot

```python
import warnings
warnings.filterwarnings('ignore')
sns.countplot(x=data['day'], data=data, palette='Set2', saturation=0.9)
```

## Violin Plot 

```python
fig,ax = plt.subplots()
fig.set_facecolor('black')
ax.set_facecolor('black')
for spines in ax.spines.values():
    spines.set_color('white')
ax.tick_params(axis='both',colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

sns.violinplot(x=data['day'], y=data['total_bill'], data=data, palette='gnuplot', hue='time', linewidth=1.5, inner='quartile')
plt.legend(loc='upper right')
plt.show()
```
